```python
import pdfplumber
```

## Read PDF FILE

```python
file_name='/home/rk/PDF_extractor/SamplePDF/pdfno2'
```

## PDF to TEXT

```python
i=0
k=''
pdf = pdfplumber.open(file_name+'.pdf')
total_pages=len(pdf.pages)
while(1):
    if i==total_pages:
        break
    first_page=pdf.pages[i]
    print(first_page.extract_text())
    k=k+first_page.extract_text()
    i=i+1
```

## Write to txt File

```python
f = open(file_name+'.txt', 'a')
f.write(k)
f.close()
```

## Read PDF in Tabular Form(Dataframe)

```python
import pandas as pd
pdf = pdfplumber.open(file_name+'.pdf')
i=0
total_pages=len(pdf.pages)
df=pd.DataFrame()
list_of_df=[]
while(1):
    if i==total_pages:
        break
    first_page=pdf.pages[i]
    try:
        table=first_page.extract_table()
    except:
        i=i+1
        continue
    tmp=pd.DataFrame(table)
    list_of_df.append(tmp)
    i=i+1

```

## Writing Complete PdF  into csv

```python
final_df=pd.concat(list_of_df,ignore_index=True)
final_df.head()
```

```python
final_df.to_csv(file_name+'.csv')
```
